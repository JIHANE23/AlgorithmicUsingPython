## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
fig=plt.figure()
plt.figure(figsize=(7,4))
df = pd.read_csv("../data/company_sales_data.csv")
df["month_number"]
df["total_profit"]
plt.plot(df.month_number,df.total_profit,label="evolution")
plt.ylabel("total profit")
plt.xlabel("month number")
plt.title("company profit per month")
plt.legend(loc="upper left")
x=np.arange(1,12,1)
plt.xticks(x)
plt.show()
 
    
    
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv("../data/company_sales_data.csv")
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
g=['ToothPaste','bathingsoap','shampoo','moisturizer','facewash','facecream']
l=[df["toothpaste"].sum(),df["bathingsoap"].sum(),df["shampoo"].sum(),df["moisturizer"].sum(),df["facewash"].sum(),df["facecream"].sum()]
ax.pie(l,labels=g,autopct='%1.1f%%')
plt.legend()
plt.show()

#exercice7
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
df=pd.read_csv("../data/company_sales_data.csv")
plt.hist(df["total_profit"],bins=[0,1,2,3,4,5])
plt.xlabel('profit range in Dollar')
plt.ylabel('actual profit in dollar')
plt.title('profit data')
plt.show()

    
    
 
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv("../data/company_sales_data.csv")
p=df["month_number"]
k=df["toothpaste"]
j=df["shampoo"]
h=df["bathingsoap"]
g=df["facewash"]
f=df["facecream"]
l=df["moisturizer"]

 



plt.stackplot(p,f,g,k,h,j,l,colors=['r','m','r','k','b','c'])
plt.xlabel("month number")
plt.ylabel("sales units in number")



plt.show()


```
